```python
def trim_context(messages: list, max_tokens: int = 900_000) -> list:
    """컨텍스트 한계 도달 시 트리밍"""
    
    total = sum(estimate_tokens(m) for m in messages)
    
    while total > max_tokens and len(messages) > 1:
        # 오래된 메시지부터 제거 (시스템 메시지 유지)
        messages.pop(1)  # 두 번째 메시지부터 제거
        total = sum(estimate_tokens(m) for m in messages)
    
    return messages
```

```python
def estimate_tokens(message: dict) -> int:
    """간단한 토큰 추정 (글자 수 / 4)"""
    content = message.get("content", "")
    if isinstance(content, list):
        content = " ".join(c.get("text", "") for c in content if isinstance(c, dict))
    return len(str(content)) // 4

# --- 테스트 ---
messages = [
    {"role": "user", "content": "시스템 메시지 역할 (유지되어야 함)"},
    {"role": "user", "content": "A" * 4000},   # ~1000 토큰
    {"role": "assistant", "content": "B" * 4000},
    {"role": "user", "content": "C" * 4000},
    {"role": "assistant", "content": "D" * 4000},
    {"role": "user", "content": "최신 메시지 (유지되어야 함)"},
]

print(f"트리밍 전 메시지 수: {len(messages)}")
print(f"총 추정 토큰: {sum(estimate_tokens(m) for m in messages):,}")

trimmed = trim_context(messages.copy(), max_tokens=2500)

print(f"\n트리밍 후 메시지 수: {len(trimmed)}")
print(f"총 추정 토큰: {sum(estimate_tokens(m) for m in trimmed):,}")
print(f"\n남은 메시지 내용 (처음 30자씩):")
for i, m in enumerate(trimmed):
    print(f"  [{i}] {m['role']}: {str(m['content'])[:30]}...")
```
